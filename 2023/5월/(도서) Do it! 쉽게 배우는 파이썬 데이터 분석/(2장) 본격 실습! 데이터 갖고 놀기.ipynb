#### **데이터 파악하기** ####
- head(): 앞부분 출력
- tail(): 뒷부분 출력
- shape: 행, 열 개수 출력
- info(): 변수 속성 출력
- describe(): 요약 통계량 출력  
  - count: 빈도
  - mean: 평균
  - std: 표준편차
  - min: 최소값
  - 25%: 1사분위수
  - 50%: 중앙값
  - 75%: 3사분위수
  - max: 최대값
  - unique: 고유값 빈도
  - top: 최빈값
  - freq:최빈값 빈도

#### **데이터 합치기** ####
- 가로로 합치기: pd.merge()
- 세로로 합치기: pd.concat()

```python
# 가로로 합치기
test1=pd.DataFrame({"id":[1,2,3,4,5],
                    "midterm":[60,80,70,90,86]})
test2=pd.DataFrame({"id":[1,2,3,4,5],
                    "final":[70,83,65,95,80]})
total=pd.merge(test1,test2,how="left",on="id")
total
```

```python
# 다른 데이터를 활용해 변수 추가하기
name=pd.DataFrame({"nclass":[1,2,3,4,5],
                   "teacher":["kim","lee","park","choi","jung"]}) # 반별 담임교사 명단
exam_new=pd.merge(exam,name,how="left",on="nclass")
exam_new
```

#### **이상치 정제하기** ####
- 이상치: 정상 범위에서 크게 벗어난 값
- 존재할 수 없는 값
- 극단적인 값
  - 상자 그림으로 극단치 기준 정하기
    - 상자 아래 세로선: 아랫수염(하위 0~25%)
    - 상자 밑면: 1사분위수(Q1, 하위 25%)
    - 상자 내 굵은 선: 2사분위수(Q2, 중앙값)
    - 상자 윗면: 3사분위수(Q3, 하위 75%)
    - 상자 위 세로선: 윗수염(하위 75~100%)
    - 상자 밖 점 표식: 극단치(Q1, Q3 밖 1.5 IQR을 벗어난 값)

```python
df=pd.DataFrame({"sex":[1,2,1,3,2,1],
                 "score":[5,4,3,4,2,6]})
print(df["sex"].value_counts())
print(df["score"].value_counts().sort_index())
```

```python
# 이상치 처리하기
# 존재할 수 없는 값
df["sex"]=np.where(df["sex"]==3,np.nan,df["sex"])
df["score"]=np.where(df["score"]>5,np.nan,df["score"])
print(df)
df.dropna(subset=["sex","score"]).groupby("sex",as_index=False).agg(mean=("score","mean"))
```
